# Ex3 : Compressible Euler simulation - Defind Problem

## Problem <a name='problem'></a>

A conservative formulation of compressible Euler equations is
\begin{align*}
\partial_t\mathbf{u}+\sum^d_{i=1}\partial_{x_i}\mathbf{F}_i(\mathbf{u}) = 0\textrm{ in }\Omega\times (0,T],
\end{align*}
where
\begin{equation*}
\mathbf{u}=\left[\begin{array}{c}
 u_1\\
 u_2\\
 u_3
 \end{array} \right] \equiv \left[\begin{array}{c}
 \rho\\
\rho u\\
E
 \end{array} \right] ,\quad \mathbf{F_1}=\left[\begin{array}{c}
 f^1_1\\
 f^2_1\\
 f^3_1
 \end{array} \right] \equiv \left[\begin{array}{c}
 \rho u\\
\rho u^2+p\\
u(E+p)
 \end{array} \right] = \left[\begin{array}{c}
u_2\\
\frac{1}{2}(3-\gamma)\frac{u_2^2}{u_1}+(\gamma-1)u_3\\
\gamma \frac{u_2}{u_1}u_3-\frac{1}{2}(\gamma-1)\frac{u_2^3}{u_1^2}
 \end{array} \right] ,
\end{equation*}
\begin{align*}
 E  =\frac{p}{(\gamma-1)}+\frac{u_2^2}{2u_1},\quad  p =(\gamma-1)\bigg(u_3-\frac{u_2^2}{2u_1}\bigg),
\end{align*}
for 1D,
\begin{align*}
\mathbf{u}=\left[\begin{array}{c}
 u_1\\
 u_2\\
 u_3\\
 u_4
 \end{array} \right] \equiv \left[\begin{array}{c}
 \rho\\
\rho u\\
\rho v\\
E
 \end{array} \right] ,\quad \mathbf{F_1} & =\left[\begin{array}{c}
 f^1_1\\
 f^2_1\\
 f^3_1\\
 f^4_1
 \end{array} \right] \equiv \left[\begin{array}{c}
 \rho u\\
\rho u^2+p\\
\rho u v\\
u(E+p)
 \end{array} \right] = \left[\begin{array}{c}
u_2\\
(\gamma-1)u_4 + \frac{3-\gamma}{2}\frac{u_2^2}{u_1} - \frac{\gamma-1}{2}\frac{u_3^2}{u_1}\\
\frac{u_2u_3}{u_1}\\
\gamma \frac{u_2u_4}{u_1} - \frac{\gamma-1}{2}\bigg(\frac{u_2^3}{u_1^2} + \frac{u_2u_3^2}{u_1^2}\bigg)
 \end{array} \right],\\
  \mathbf{F_2} & =\left[\begin{array}{c}
 f^1_2\\
 f^2_2\\
 f^3_2\\
 f^4_2
 \end{array} \right] \equiv \left[\begin{array}{c}
 \rho v\\
 \rho u v\\
\rho v^2+p\\
v(E+p)
 \end{array} \right] = \left[\begin{array}{c}
u_3\\
\frac{u_2u_3}{u_1}\\
(\gamma-1)u_4 + \frac{3-\gamma}{2}\frac{u_3^2}{u_1} - \frac{\gamma-1}{2}\frac{u_2^2}{u_1}\\
\gamma\frac{u_3u_4}{u_1} - \frac{\gamma-1}{2}\bigg(\frac{u_2^2u_3}{u_1^2} + \frac{u_3^3}{u_1^2}\bigg)
 \end{array} \right],
\end{align*}
\begin{align*}
E = \frac{p}{\gamma-1}+\frac{\rho}{2}(u^2+v^2),\quad
 p =(\gamma-1)\bigg(u_4-\frac{u_2^2+u_3^2}{2u_1}\bigg),
 \end{align*}
 for 2D, and 
 \begin{align*}
\mathbf{u}=\left[\begin{array}{c}
 u_1\\
 u_2\\
 u_3\\
 u_4\\
 u_5
 \end{array} \right] \equiv \left[\begin{array}{c}
 \rho\\
\rho u\\
\rho v\\
\rho w\\
E
 \end{array} \right] ,\quad \mathbf{F_1} & =\left[\begin{array}{c}
 f^1_1\\
 f^2_1\\
 f^3_1\\
 f^4_1\\
 f^5_1
 \end{array} \right] \equiv \left[\begin{array}{c}
 \rho u\\
\rho u^2+p\\
\rho u v\\
\rho u w\\
u(E+p)
 \end{array} \right] = \begin{bmatrix}
u_2 \\
\frac{u_2^2}{u_1} + (\gamma - 1) \left( u_5 - \frac{1}{2} \frac{u_2^2 + u_3^2 + u_4^2}{u_1} \right) \\
\frac{u_2 u_3}{u_1} \\
\frac{u_2 u_4}{u_1} \\
\left( u_5 + (\gamma - 1) \left( u_5 - \frac{1}{2} \frac{u_2^2 + u_3^2 + u_4^2}{u_1} \right) \right) \frac{u_2}{u_1}
\end{bmatrix},\\
  \mathbf{F_2} & =\left[\begin{array}{c}
 f^1_2\\
 f^2_2\\
 f^3_2\\
 f^4_2\\
 f^5_2
 \end{array} \right] \equiv \left[\begin{array}{c}
 \rho v\\
 \rho u v\\
\rho v^2+p\\
\rho v w\\
v(E+p)
 \end{array} \right] = \begin{bmatrix}
u_3 \\
\frac{u_2 u_3}{u_1} \\
\frac{u_3^2}{u_1} + (\gamma - 1) \left( u_5 - \frac{1}{2} \frac{u_2^2 + u_3^2 + u_4^2}{u_1} \right) \\
\frac{u_3 u_4}{u_1} \\
\left( u_5 + (\gamma - 1) \left( u_5 - \frac{1}{2} \frac{u_2^2 + u_3^2 + u_4^2}{u_1} \right) \right) \frac{u_3}{u_1}
\end{bmatrix},\\
 \mathbf{F_3} & =\left[\begin{array}{c}
 f^1_3\\
 f^2_3\\
 f^3_3\\
 f^4_3\\
 f^5_3
 \end{array} \right] \equiv \left[\begin{array}{c}
 \rho w\\
 \rho u w\\
\rho v w\\
\rho w^2 +p\\
v(E+p)
 \end{array} \right] = \begin{bmatrix}
u_4 \\
\frac{u_2 u_4}{u_1} \\
\frac{u_3 u_4}{u_1} \\
\frac{u_4^2}{u_1} + (\gamma - 1) \left( u_5 - \frac{1}{2} \frac{u_2^2 + u_3^2 + u_4^2}{u_1} \right) \\
\left( u_5 + (\gamma - 1) \left( u_5 - \frac{1}{2} \frac{u_2^2 + u_3^2 + u_4^2}{u_1} \right) \right) \frac{u_4}{u_1}
\end{bmatrix},
\end{align*}
\begin{align*}
E = \frac{p}{\gamma-1}+\frac{\rho}{2}(u^2+v^2+w^2),\quad
 p =(\gamma-1)\bigg(u_5-\frac{u_2^2+u_3^2+u_4^2}{2u_1}\bigg),
 \end{align*} 
 for 3D. Here, $\rho$ is a density, $u,v,w$ are $x,y,z$-direction velocities, $E$ is a total energy, $p$ is the pressure and $\gamma=1.4$.


In this example, we provide a accuracy test case for the three-dimensional problem. We consider the cube $\Omega = (0,1)^3$ with periodic boundary conditions, and use the initial conditions, given by
\begin{align}
\rho & = 1+0.2 \sin{\big(2\pi(x+y+z)\big)},\\
u & = 1, \quad v=-1/2,\quad w=1,\\
p & = 1.
\end{align}
The exact solution to the Euler equations is known analytically in this case. Velocity and pressure remain constant in time, and the density at time $t$ is given by
\begin{equation}
\rho = 1+0.2 \sin{\big(2\pi(x+y+z-t(u+v+w))\big)}.
\end{equation}

## DGLM formulation <a name='dg_form'></a>

By the nodal DG approximation, on all K elements, 

A: 

$$ M^k \frac{d}{dt} u^k_0 - (r_x D^T_r + s_x D^T_s + t_x D^T_t) M^k v_x u^k_0 - (r_y D^T_r + s_y D^T_s + s_y D^T_t) M^k v_y u^k_0 - (r_z D^T_r + s_z D^T_s + s_z D^T_t) M^k v_z u^k_0 + \sum_\gamma \tau_\gamma J_s M_\gamma u^k_0 $$

$$
\begin{aligned}
& \text{(RHS)} \\
\Rightarrow&
(r_x M^{-1} D^T_r M + s_x M^{-1} D^T_s M + t_x M^{-1} D^T_t M) v_x u^k_0 
+ (r_y M^{-1} D^T_r M + s_y M^{-1} D^T_s M + t_y M^{-1} D^T_t M) v_y u^k_0 
+ (r_z M^{-1} D^T_r M + s_z M^{-1} D^T_s M + t_z M^{-1} D^T_t M) v_z u^k_0 
- \sum_\gamma \tau_\gamma J^{-1} M^{-1} J_s M_\gamma u^k_0 \\
=&
(r_x D'_r + s_x D'_s + t_x D'_t) v_x u^k_0
+ (r_y D'_r + s_y D'_s + t_y D'_t) v_y u^k_0
+ (r_z D'_r + s_z D'_s + t_z D'_t) v_z u^k_0
- \sum_\gamma \tau_\gamma J^{-1} M^{-1} J_s M_\gamma u^k_0 \\
\end{aligned}
$$

where $D'_r = M^{-1} D^T_r M$, $D'_s = M^{-1} D^T_s M$, and $D'_t = M^{-1} D^T_t M$

B: 

$$
\begin{aligned}
& \text{(RHS)} \\
& \sum_\gamma - J^{-1} M^{-1} J_s M_\gamma \left( n_x v_x u^k_b + n_y v_y u^k_b + n_z v_z u^k_b \right)
+  \tau_\gamma J^{-1} M^{-1} J_s M_\gamma u^k_b \\
=& - J^{-1} \sum_\gamma M^{-1} M_\gamma J_s \left( n_x v_x u^k_b + n_y v_y u^k_b + n_z v_z u^k_b \right)
- \tau_\gamma M^{-1} M_\gamma J_s u^k_b \\
=& - J^{-1} \sum_\gamma M^{-1} M_\gamma J_s \left( n_x v_x u^k_b + n_y v_y u^k_b + n_z v_z u^k_b - \tau_\gamma u^k_b \right) \\
=& - \text{LIFT}@\text{Fscale}. \left( n_x v_x u^k_b + n_y v_y u^k_b + n_z v_z u^k_b - \tau_\gamma u^k_b \right)
\end{aligned}
$$

where $\text{LIFT}=M^{-1} M_\gamma$ and $\text{Fscale}=J_s/J^k$.

R: 

$$ r = 0 $$

If the Dirichlet boundary conditions are applied

$$
\text{(RHS)}
- J^{-1} M^{-1} J_s M_s \left( n_x v_x \mathbf{g} + n_y v_y \mathbf{g} + n_z v_z \mathbf{g} \right)
+ \sum_\gamma \tau_\gamma J^{-1} M^{-1} J_s M_s \mathbf{g}
$$

where $\mathbf{g}$ is the Dirichlet boundary conditions.

So we have the explicit DGLM evolution

$$ 
\begin{aligned}
\frac{d}{dt} u^k_0 
&= A + B + R \\
&= (r_x D'_r + s_x D'_s + s_x D'_t) v_x u^k_0 + (r_y D'_r + s_y D'_s + s_y D'_t) v_y u^k_0 + (r_z D'_r + s_z D'_s + t_z D'_t) v_z u^k_0
- \text{LIFT}@\text{Fscale}. \left[ n_x v_x u^k_b + n_y v_y u^k_b  + n_z v_z u^k_b - \tau_\gamma \left( u^k_b - u^k_0 \right) \right] \\
\end{aligned}
$$

On the Dirichlet boundary,

$$
= (r_x D'_r + s_x D'_s + t_x D'_t) v_x u^k_0 + (r_y D'_r + s_y D'_s + t_y D'_t) v_y u^k_0 + (r_z D'_r + s_z D'_s + t_z D'_t) v_z u^k_0
- \text{LIFT}@\text{Fscale}. \left[ n_x v_x \mathbf{g} + n_y v_y \mathbf{g} + n_z v_z \mathbf{g} - \tau_\gamma \left( \mathbf{g} - u^k_0 \right) \right]
$$

We have to update $u^k_b$

$$ u_b = \frac{u^-_0 + u^+_0}{2} $$
